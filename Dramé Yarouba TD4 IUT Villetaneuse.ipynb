### Exercice 1 : Test de compréhension*

#### Question 1

Quelle est la différence entre *définition* et *appel* d'une fonction ?

Défintion : On donne à la fonction une séquence d'instruction qu'elle va exécuter quand on va l'appeller

Appel : On demande à la fonction d'exécuter la séquence d'instruction qn'on lui a définit

#### Question 2

Qu'est-ce que l'*en-tête* d'une fonction ? Qu'est-ce que le *corps* d'une fonction ?

En-tête : La première ligne quand on défint une fonction

Corps : La séquence d'instruction donnée à la fonction

#### Question 3

Qu'est-ce qu'un *programme appelant* ?

Programme appelant : Ligne de code qui utilise l'en-tête d'une fonction avec des valeurs données aux paramétres entre-parenthèses qui va appeller la fonction pour demander son exécution avec les valeurs et paramétres données

Oui, dans les deux cas, les parenthèses sont obligatoires pour la définir et pour l'appeler même s'il n'y a aucun paramètre.

### Exercice 5 : Comparer deux nombres*

Écrire la fonction `compare()` qui reçoit deux nombres entiers en paramètre `a` et `b` et affiche un message adapté selon que `a` est supérieur, inférieur ou égal à `b`. 

````python
#-------------------------#
def compare(a,b) :
    if a > b :
        print (a, "est supérieur à" ,b)
    elif a < b :
        print (a, "est inférieur à" ,b)
    else :
        print (a, "est égal à", b)

compare(6,2)
compare(2,6)
compare(6,6)        
#-------------------------#


````

### Exercice 6 : Addition unaire*

Un nombre positif ou nul est représenté en base unaire par autant de symboles 0 que la valeur du nombre. Ainsi, la représentation unaire de 5 est : `"00000"`.

1. Définir la fonction `affiche_unaire` prenant en paramètre un nombre et affichant ce nombre en unaire.
2. Définir la fonction `addition_unaire` prenant en paramètre deux nombres et affichant ces deux nombres en base unaire ainsi que leur somme. Par exemple, l'appel de la fonction avec les valeurs 7 et 5 doit donner l'affichage suivant :
```
0000000
+
00000
=
000000000000
```

````python
#-------------------------#
def affiche_unaire (nb) :
    n = 0
    s = ""
    while n < nb :
        s += "0"
        n += 1
    print (s)



def addition_unaire (nb1, nb2) :
    affiche_unaire(nb1)
    print("+")
    affiche_unaire(nb2)
    print("=")
    print(affiche_unaire(nb1 + nb2))
          
addition_unaire(7,5)          
#-------------------------#


````
